```python
# 查找当前目录及子目录下所有文件
!find .
```

```python
# 查找当前目录及子目录下所有 .ipynb 后缀的文件
!find . -name "*.ipynb"
```

```python
# 忽略大小写
!find . -iname "*.ipynb"
```

```python
# 查找当前目录及子目录下所有不是 .ipynb 后缀的文件
!find . ! -name "*.ipynb"
```

```python
# 向下查找 1 层深度
!find . -maxdepth 1 
```

```python
# 根据文件类型进行搜索
!find . -type f
```

类型参数列表：

- f 普通文件
- l 符号连接
- d 目录
- c 字符设备
- b 块设备
- s 套接字
- p Fifo

```python
# 查找上层目录并多匹配
!find .. -name "*.py" -o -name "*.pdf"
```

```python
# 匹配文件路径或者文件
!find . -path "*ipynb*"
```

```python
# 基于正则表达式匹配文件路径
!find . -regex ".*ipynb"
```

```python
# 搜索大于 1k 的文件
!find . -type f -size +1k
```

```python
# 搜索小于 100k 的文件
!find . -type f -size -100k
```

```python
# 搜索等于 100k 的文件
!find . -type f -size 100k
```

文件大小单元：

- b —— 块（512字节）
- c —— 字节
- w —— 字（2字节）
- k —— 千字节
- M —— 兆字节
- G —— 吉字节
